```python
from sklearn.preprocessing import StandardScaler

# Columns used for clustering in your original df
features = ['danceability', 'energy', 'loudness', 'speechiness', 
            'acousticness', 'instrumentalness', 'liveness', 
            'valence', 'tempo', 'duration_ms']

scaler = StandardScaler()
scaler.fit(df[features])  # fit only on these numeric features

# Now transform uploaded song with same columns
df_scaled = scaler.transform(df[features])
```

```python
# K-Means Clustering with Elbow Method and Silhouette Score
# -------------------------------
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# ---------- 1. Elbow Method ----------
wcss = []
# Test cluster sizes from 1 to 30
for i in range(1, 31):
    kmeans_value = KMeans(n_clusters=i, random_state=0, n_init='auto', max_iter=3000)
    kmeans_value.fit(df_scaled)  # df_scaled = scaled music features
    wcss.append(kmeans_value.inertia_)

# Plot WCSS to find optimal k
plt.figure(figsize=(8,5))
plt.plot(range(1, 31), wcss, marker='o')
plt.xlabel('Number of clusters (k)')
plt.ylabel('WCSS (Inertia)')
plt.title('Elbow Method for Optimal k')
plt.show()
```

```python
# ---------- 2. Fit K-Means with chosen k ----------
optimal_k = 5  # Replace with the "elbow" value you choose
kmeans_value = KMeans(n_clusters=optimal_k, random_state=0, n_init='auto', max_iter=3000)
kmeans_value.fit(df_scaled)

# Assign cluster labels to the dataset
df['Cluster'] = kmeans_value.predict(df_scaled)[0]
print(df.head())
```

```python
# ---------- 3. Silhouette Score ----------
s_score = silhouette_score(df_scaled, kmeans_value.labels_)
print("Silhouette Score:", s_score)
```

```python
# PCA 2D Visualization of K-Means Clusters
# -------------------------------
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Reduce dimensions to 2 for visualization
pca = PCA(n_components=2, random_state=0)
df_pca = pca.fit_transform(df_scaled)

# Plot clusters
plt.figure(figsize=(8,6))
scatter = plt.scatter(df_pca[:,0], df_pca[:,1], c=df['Cluster'], cmap='tab10', alpha=0.6)
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.title('K-Means Clusters Visualization (PCA 2D)')
plt.legend(*scatter.legend_elements(), title="Clusters")
plt.show()
```

```python
cluster_profile = df.groupby('Cluster').mean()
print(cluster_profile)
```

```python
cluster_labels = {
    0: 'Instrumental / Acoustic',
    1: 'Mainstream / Party',
    2: 'Happy / Dance',
    3: 'Vocal / Speech-heavy',
    4: 'Chill / Relaxing'
}
df['Cluster_Label'] = df['Cluster'].map(cluster_labels)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Select only numeric columns
numeric_df = df.select_dtypes(include=np.number)

# Compute mean of numeric features per cluster
cluster_profile = numeric_df.groupby(df['Cluster']).mean()

# Plot heatmap safely
plt.figure(figsize=(12,6))
sns.heatmap(cluster_profile, annot=True, cmap='YlGnBu', fmt=".2f")
plt.title("Cluster Feature Profile Heatmap")
plt.show()
```
